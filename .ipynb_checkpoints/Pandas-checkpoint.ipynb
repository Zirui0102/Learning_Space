````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
#dataframe..head() returns the first few rows (the “head” of the DataFrame).
#.info() shows information on each of the columns, such as the data type and number of missing values.
#.shape returns the number of rows and columns of the DataFrame.
#.describe() calculates a few summary statistics for each column.

#Three components of Dataframe objects.
#.values: A two-dimensional NumPy array of values.
#.columns: An index of columns: the column names.
#.index: An index for the rows: either row numbers or row names.
````

**Functions**
1. `data.sort_values("", ascending=)`: sorting
2. `data[""]`: subsetting columns
3. `data[""].isin([""])`: search categorical variables
4. `data[""] = ...`: adding a new column
5. `data.drop("", axis=)`: drop a column
6. `data[""].max()`: summary statistics; other call: `.median()`, `.mode()`, `.min()`, `.max()`, `.var()`, `.std()`, `.sum()`, `.quantile()`
7. `data[""].cumsum()`: cumulative sum; other call: `.cummax()`, `.cummin()`, `.cumprod()`
8. `data.drop_duplicates(subset=)`: dropping duplicate names
9. `data[""].value_counts(sort=)`: count values
10. `data.groupby("")[""].mean()`: grouped summaries
11. `data.pivot_table(value=, index=, columns=, fill_value=)`: group by to pivot table
12. `data.set_index("")`: setting a column as the index
13. `data.columns` & `data.index`: column and row numbers
14. `data.reset_index(drop=True)`: removing an index; `drop=True`, completely drop the index column
15. `data.loc[[]]`: subset the outer level with a list
16. `data.sort_index(level=, ascending=)`: sorting by index values
17. `data.loc[:,:]`: slicing the index
18. `data.iloc[:,:]`: subsetting by row/column number
19. `data.mean(axis="index"/"columns")`: the axis argument
20. `data[""].hist()`: visualization by histogram
21. `data.plot(kind="bar"/"line"/"scatter")`: bar/line/scatter chart
22. `data.isna()`: detecting missing values
23. `data.isna().any()`
24. `data.isna().sum()`: counting missing values
25. `data.dropna()`: removing missing values
26. `data.fillna(0)`: replacing missing values
27. ```python
    list_of_dicts = [
        {"name": "Ginger", "breed": "Dachshund", "height_cm": 22},
        {"name": "Terry", "breed": "Dalmatian", "height_cm": 59}
    ]
    new_data = pd.DataFrame(list_of_dicts)  # list of dictionaries - by row
    ```
28. ```python
    dict_of_lists = {
        "name": ["Ginger", "Terry"],
        "breed": ["Dachshund", "Dalmatian"],
        "height_cm": [22, 59]
    }
    new_data2 = pd.DataFrame(dict_of_lists)  # dictionary of lists - by column
    ```
29. `pd.read_csv()`: reading CSVs
30. `.to_csv()`: writing CSVs

**1. Sorting and subsetting**

````python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)
````

````python
data.head()
````

````python
#Sorting
data.sort_values("Age")
````

````python
#Sorting in descending order
data.sort_values("Age", ascending=False)
````

````python
#Sorting by multiple variables
data.sort_values(["Age", "Ticket"])
````

````python
#Sorting by multiple variables
data.sort_values(["Age", "Ticket"], ascending=[False, True])
````

````python
#Subsetting columns
data["Name"]
````

````python
#Subsetting multiple columns
data[["Name", "Survived"]]
````

````python
#Subsetting rows
data[data["Age"] >= 70]
````

````python
#Subsetting based on text data
data[data["Sex"] == "male"]
````

````python
#Subsetting based on multiple conditions
is_male = data["Sex"] == "male"
is_70 = data["Age"] > 70
data[is_male & is_70]
````

````python
data[(data["Sex"] == "male") & (data["Age"] > 70)]
````

````python
#Subsetting using .isin(): search categorical variables
is_male = data["Sex"].isin(["male"])
data[is_male]
````

**3. New columns**

````python
#adding a new column
data["PassengerId/10"] = data["PassengerId"] / 10
data
````

````python
#drop a column
data.drop('PassengerId/10', axis=1)
````

**4. Summary statistics**

````python
#summarizing numerical data: .mean()
#other call: .median(), .mode(), .min(), .max()
#.var(), .std(), .sum(), .quantile()
data["Age"].max()
````

````python
#the .agg() method
def pct30(column):
    return column.quantile(0.3)
data[["Age", "Fare"]].agg(pct30)
````

````python
data["Age"].agg([pct30, np.median])
````

````python
def pct40(column):
    return column.quantile(0.4)
data[["Age", "Fare"]].agg([pct30, pct40])
````

````python
#cumulative sum: .cumsum()
#other calls: .cummax(), .cummin(), .cumprod() 
data["Age"].cumsum()
````

**5. Counting categorical variables**

````python
#dropping duplicate names
data.drop_duplicates(subset="Ticket")
````

````python
#dropping duplicate names
data1 = data.drop_duplicates(subset=["Ticket", "Name"])
data1
````

````python
# .value_counts
data1["Ticket"].value_counts(sort=True)
````

````python
data1["Ticket"].value_counts(normalize=True)
````

**6. Grouped summary statistics**

````python
#Summaries by group
data[data["Sex"] == "male"]["Age"].mean()
````

````python
#Grouped summaries
data.groupby("Sex")["Age"].mean()
````

````python
#multiple grouped summaries
data.groupby("Sex")["Age"].agg([min, max, sum])
````

````python
#grouping by multiple variables
data.groupby(["Sex", "Survived"])["Age"].mean()
````

**7. Pivot tables**

````python
#group by to pivot table
data.pivot_table(values="Age", index="Sex")
````

````python
#Different statistics
data.pivot_table(values="Age", index="Sex", aggfunc=[np.median, np.mean])
````

````python
#Pivot on two variables
data.pivot_table(values="Age", index="Sex", columns="Survived")
````

````python
data.groupby(["Sex", "Survived"])["Age"].mean()
````

````python
#Filling missing values in pivot tables, summary statistics using margins=True.
data.pivot_table(values="Age", index="Name", columns="Survived", fill_value=0, margins=True)
````

**7. Explicit indexes**

````python
#.columns and .index
data.columns
````

````python
#row numbers
data.index
````

````python
#setting a column as the index
data1 = data.set_index("Name")
data1.head()
````

````python
#removing an index; drop=True, completely drop the index column
data1.reset_index(drop=True, inplace=True)
data1.head()
````

````python
data.head()
````

````python
#indexes make subsetting simpler
data[data["Name"].isin(["Braund, Mr. Owen Harris"])]
````

````python
data1.loc[["Braund, Mr. Owen Harris"]]
````

````python
#multi-level indexes a.k.a hierarchical indexes
data2 = data.set_index(["Name", "Age"])
data2.head()
````

````python
#subset the outer level with a list
data2.loc[["Braund, Mr. Owen Harris", "Heikkinen, Miss. Laina"]]
````

````python
data2.loc[[("Braund, Mr. Owen Harris", 22)]]
````

````python
#sorting by index values
data2.sort_index(level=["Age", "Name"], ascending=[False, False])
````

**8. Slicing and subsetting with .loc and iloc**

````python
#sort the index before uyou slice
data3 = data.set_index(["Name", "Age"]).sort_index()
data3.head()
````

````python
#slicing the outer index level
data3.loc["Abbing, Mr. Anthony":"Abbott, Mrs. Stanton (Rosa Hunt)"]
````

````python
#slicing the inner index levels
data3.loc[("Abbing, Mr. Anthony", 42):("Abbott, Mr. Rossmore Edward", 16)]
````

````python
#slicing columns
data3.loc[:, :"Ticket"]
````

````python
#subsetting by row/column number
data3.iloc[2:5, 1:4]
````

**9. Working with pivot tables**

````python
data1 = data.pivot_table("Age", index=["Name", "PassengerId"], columns="Sex")
data1.head()
````

````python
#slicing the pivot table
data1.loc["Abbing, Mr. Anthony":"Abbott, Mrs. Stanton (Rosa Hunt)"]
````

````python
#the axis argument
data1.mean(axis="index")
````

````python
#the axis argument
data1.mean(axis="columns")
````

**10. Visualizing your data**

````python
data["Age"].hist()
plt.show()
````

````python
data["Age"].hist(bins=20)
plt.show()
````

````python
avg_age = data.groupby("Sex")["Age"].mean()
avg_age.plot(kind="bar", width=0.3, title="Mean Age by Sex")
````

````python
data.plot(x="Age", y="PassengerId", kind="line", rot=45)
plt.show()
````

````python
data.plot(x="Age", y="PassengerId", kind="scatter", rot=45)
plt.show()
````

````python
data[data["Sex"]=="female"]["Age"].hist(alpha=0.7)
data[data["Sex"]=="male"]["Age"].hist(alpha=0.7)
plt.legend(["female", "male"])
plt.show()
````

**11. Missing values**

````python
#detecting missing values
data.isna()
````

````python
data.isna().any()
````

````python
#counting missing values
data.isna().sum()
````

````python
#plotting missing values
data.isna().sum().plot(kind="bar")
plt.show()
````

````python
#removing missing values
data.dropna().head()
````

````python
#replacing missing values
data.fillna(0).head()
````

**12. Creating DataFrames**

````python
#list of dictionaries - by row
list_of_dicts = [
    {"name": "Ginger", "breed": "Dachshund", "height_cm": 22},
    {"name": "Terry", "breed": "Dalmatian", "height_cm": 59}
]
new_data = pd.DataFrame(list_of_dicts)
new_data
````

````python
#Dictionary of lists - by column (key=column name; value=list of column values)
dict_of_lists = {
    "name": ["Ginger", "Terry"],
    "breed": ["Dachshund", "Dalmatian"],
    "height_cm": [22, 59]
}

new_data2 = pd.DataFrame(dict_of_lists)
new_data2
````
