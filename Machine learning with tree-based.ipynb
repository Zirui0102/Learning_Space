**4. Generalization Error**

```python
#Generalization Error of f_hat: does f_hat generalize well on unseen data?
#Generalization Error of f_hat = bias^2 + variance + irreducible error
#Bias: error term that tells you, on average, how much f_hat not equal to f (underfitting -> high bias)
#Variance: how much f_hat is inconsistent over different training sets (overfitting -> high variance)
#Model complexity: sets the flexibility of f_hat. Example: maximum tree depth, minimum samples per leaf.
#Model complexity increases, the variance increases while the bias decreases. Model complexity decreases, the variance decreases while the bias increases (Bias-Variance Tradeoff).
#Your goal is to find the model complexity that achieves the lowest generalization error.
```

**5. Estimating the Generalization Error**

```python
#Solution -> Cross-Validation (CV): K-Fold CV, Hold-Out CV
```

```python
#If f_hat suffers from high variance: CV error of f_hat > training set error of f_hat
#f_hat is said to overfit the training set. To remedy overfitting:
#decrease model complexity; 
#for ex: decrease max depth, increase min samples per leaf.
#gather more data.
```

```python
#If f_hat suffers from high bias: CV error of f_hat is similar to training set error of f_hat >> desired error
#f_hat is said to underfit the training set. To remedy overfitting:
#increase model complexity; 
#for ex: increase max depth, decrease min samples per leaf.
#gather more relevant features.
```

```python
#Task: predict the mpg consumption of a car given the chosen features.
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import cross_val_score

X = data.drop(['mpg'], axis=1)
y = data['mpg']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3,
                                                   random_state=123)
dt = DecisionTreeRegressor(max_depth=4,
                           min_samples_leaf=0.14, 
                           random_state=123)

MSE_CV = - cross_val_score(dt, X_train, y_train, cv=10,
                           scoring='neg_mean_squared_error',
                           n_jobs = -1)

dt.fit(X_train, y_train)

y_predict_train = dt.predict(X_train)
y_predict_test = dt.predict(X_test)

#CV MSE
print('CV MSE: {:.2f}'.format(MSE_CV.mean()))
#Training set MSE
print('Train MSE: {:.2f}'.format(MSE(y_train, y_predict_train)))
#Test set MSE
print('Test MSE: {:.2f}'.format(MSE(y_test, y_predict_test)))
```

**6. Ensemble Learning**

**Ensemble Learning**
* Train different models on the same dataset.
* Let each model make its predictions.
* Meta-model: aggregates predictions of individual models.
* Final prediction: more robust and less prone to errors.
* Best results: models are skillful in different ways.

```python
#Auto-mpg dataset
data = pd.read_csv('C:/Users/89751/OneDrive/Desktop/Learning_space/data_indian_liver_patient.csv')
# data.replace('?', np.nan, inplace=True)  # Replace '?' with NaN
data.dropna(inplace=True)  # Drop rows with NaN values
data.head()
```

```python
#Label encoding
import numpy as np
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['gender'] = label_encoder.fit_transform(data['gender'])
data.head()
```

```python
X = data.drop(['is_patient'], axis=1)
y = data['is_patient']
```

```python
#import functions to compute accuracy and split data
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

#import models, including votingclassifier meta-model
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import VotingClassifier
```

```python
X = data.drop(['is_patient'], axis=1)
y = data['is_patient']

#set seed for reproducibility
SEED = 123

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)
#instantiate individual classifiers
lr = LogisticRegression(random_state=SEED)
knn = KNN(n_neighbors=27)
dt = DecisionTreeClassifier(min_samples_leaf=0.13, random_state=SEED)

#define a list called classifier that contains the tuples (classifier_name, classifier)
classifiers = [('Logistic regression', lr),
               ('K Nearest Neighbours', knn),
               ('Classification Tree', dt)]
```

```python
#iterate over the defined list of tuples containing the classifiers
for clf_name, clf in classifiers:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print('{:s} : {:.3f}'.format(clf_name, accuracy_score(y_test, y_pred)))
```

```python
# Instantiate a VotingClassifier vc
vc = VotingClassifier(estimators=classifiers)     

# Fit vc to the training set
vc.fit(X_train, y_train)   

# Evaluate the test set predictions
y_pred = vc.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print('Voting Classifier: {:.3f}'.format(accuracy))
```

**7. Bagging**

**Bagging**: Bootstrap Aggregation.
* One algorithms; different subsets of the training set.
* Uses a technique known as the bootstrap.
* Reduces variance of individual models in the ensemble.
* Classification (BaggingClassifier) - aggregates predictions by majority voting; Regression (Baggingregressor) - aggregates predictions through averaging.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 1

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)
```

```python
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=0.13, random_state=SEED)
bc = BaggingClassifier(estimator=dt, n_estimators=300, n_jobs=-1)

bc.fit(X_train, y_train)
y_pred = bc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of Bagging Classifier: {:.3f}'.format(accuracy))
```

**8. Out of Bag Evaluation**

**Out Of Bag (ODB) instances**
* On average, for each model, 63% of the training instances are sampled.
* The remaining 37% constitute the OOB instances, which can be used to estimate the performance of the ensemble (without cross validation)
* In scikit-learn, the OOB-score corresponds to the accuracy for classifiers and the r-squared score for regressors.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 1

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)
```

```python
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=0.16, random_state=SEED)
bc = BaggingClassifier(estimator=dt, n_estimators=300, oob_score=True, n_jobs=-1)

bc.fit(X_train, y_train)
y_pred = bc.predict(X_test)

test_accuracy = accuracy_score(y_test, y_pred)
ood_accuracy = bc.oob_score_

print('Test set accuracy: {:.3f}'.format(test_accuracy))
print('OOB accuracy: {:.3f}'.format(ood_accuracy))
```

**10. AddBoost**

**Boosting**
* Boosting: ensemble method combining several weak learners to form a strong learner. 
* Weak learner: model doing slightly better than random guessing (e.g., Decision stump)
* Train an ensemble of predictors sequentially.
* Each predictor tries to correct its predecessor.
* AddBoost (Adaptive Boosting): Each predictor pays more attention to the instances wrongly predicted by its predecessor. Achieved by changing the weights of training instances. Each predictor is assigned a coefficient α. α depends on the predictor's training errors.
* Learning rate (eta): 0 < eta < 1; it used to shrink the coefficient alpha of a trained predictor. (a smaller value of eta should be compensated by a greater number of estimators.)

**AddBoost**
* Classification: aggregates predictions by majority voting; (AdaBoostClassifier in scikit-learn).
* Regression: aggregates predictions through averaging; (AdaBoostRegressor in scikit-learn).

**11. Gradient Boosting (GB)**

**Gradient Boosted Trees**
* Sequential correction of predecessor's errors.
* Does not tweak the weights of training instances.
* Gradient boosting minimizes a loss function like MSE or log loss while AdaBoost focuses on instances with high error by adjusting their sample weights adaptively.
* Fit each predictor is trained using its predecessor's residual errors as labels.
* Gradient Boosted Trees: a CART is used as a base learner.
* Shrinkage: refers to the fact that the prediction of each tree emsemble is shrinked after it is multiplied by a learning rate eta (apply to avoid overfitting).

**GB**
* Classification: aggregates predictions by majority voting; (GradientBoostingRegressor in scikit-learn).
* Regression: aggregates predictions through averaging; (GradientBoostingClassifier in scikit-learn).

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE

SEED = 1

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)
```

```python
gbt = GradientBoostingRegressor(n_estimators=300, max_depth=1, random_state=SEED)

gbt.fit(X_train, y_train)

y_pred = gbt.predict(X_test)

rmse_test = MSE(y_test, y_pred)**(1/2)

print('Test set RMSE: {:.2f}'.format(rmse_test))
```

**12. Stochastic Gradient Boosting (SGB)**

**SGB**
* Each tree is trained on a random subset of rows of the training data.
* The sampled instances (40%-80% of the training set) are sampled without replacement.
* Features are sampled (without replacement) when choosing split points.
* Result: further ensemble diversity.
* Effect: adding further variance to the ensemble of trees.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE

SEED = 1

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)
```

```python
sgbt = GradientBoostingRegressor(max_depth=1,
                                subsample=0.8,  #each tree to sample 80% of the data for training
                                max_features=0.2,  #each tree uses 20% of available features to perform the best-split
                                n_estimators=300,
                                random_state=SEED)

sgbt.fit(X_train, y_train)

y_pred = sgbt.predict(X_test)

rmse_test = MSE(y_test, y_pred)**(1/2)

print('Test set RMSE: {:.2f}'.format(rmse_test))
```

**13. Tuning a CART's hyperparameters**

* Parameters: learned from data (e.g., split-point of a node, split-feature of a node)
* Hyperparameters: not learned from data, set prior to training (e.g., max_depth, min_samples_leaf)
* Hyperparameter tuning: search for a set of optimal hyperparameters for a learning algorithm.

**Grid search cross validation**
* Manually set a grid of discrete hyperparameter values.
* Set a metric for scoring model performance.
* Search exhaustively through the grid.
* For each set of hyperparameters, evaluate each model's CV score.
* The optimal hyperparameters are those of the model achieving the best CV score.

```python
#Auto-mpg dataset
data = pd.read_csv('C:/Users/89751/OneDrive/Desktop/Learning_space/data_indian_liver_patient.csv')
# data.replace('?', np.nan, inplace=True)  # Replace '?' with NaN
data.dropna(inplace=True)  # Drop rows with NaN values
data.head()
```

```python
label_encoder = LabelEncoder()
data['gender'] = label_encoder.fit_transform(data['gender'])
data.head()
```

```python
X = data.drop(['is_patient'], axis=1)
y = data['is_patient']
```

```python
from sklearn.tree import DecisionTreeClassifier

SEED = 1

dt = DecisionTreeClassifier(random_state=SEED)

print(dt.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=SEED)

params_dt = {
            'max_depth':[3, 4, 5, 6],
            'min_samples_leaf': [0.04, 0.06, 0.08],
            'max_features': [0.2, 0.4, 0.6, 0.8]
            }
grid_dt = GridSearchCV(estimator=dt,
                      param_grid=params_dt,
                      scoring='accuracy',
                      cv=10,  #use 10-fold stratified cross-validation for model evaluation.
                      n_jobs=-1)

grid_dt.fit(X_train, y_train)

best_hyperparams = grid_dt.best_params_
print('Best hyperparameters:\n', best_hyperparams)
```

```python
best_CV_score = grid_dt.best_score_
print('Best CV accuract: {:.2f}'.format(best_CV_score))
```

```python
#extract best model from 'grid_dt'
best_model = grid_dt.best_estimator_

test_acc = best_model.score(X_test, y_test)

print("Test set accuracy of best model: {:.3f}".format(test_acc))
```
